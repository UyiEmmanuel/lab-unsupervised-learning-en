```python
# Calculate the cumulative percentage of total spending per customer
total_spending = data.sum(axis=1)
pareto_data = pd.DataFrame({'Customer': np.arange(1, len(total_spending)+1), 'Total_Spending': total_spending})
pareto_data = pareto_data.sort_values(by='Total_Spending', ascending=False)
pareto_data['Cumulative_Percentage'] = pareto_data['Total_Spending'].cumsum() / pareto_data['Total_Spending'].sum() * 100

# Plot the Pareto principle
plt.figure(figsize=(10, 6))
plt.plot(pareto_data['Customer'], pareto_data['Cumulative_Percentage'], marker='o')
plt.axhline(y=80, color='r', linestyle='--')
plt.title('Pareto Principle')
plt.xlabel('Number of Customers')
plt.ylabel('Cumulative Percentage of Total Spending')
plt.show()
```

# Bonus Challenge 2 - Changing K-Means Number of Clusters

As we mentioned earlier, we don't need to worry about the number of clusters with DBSCAN because it automatically decides that based on the parameters we send to it. But with K-Means, we have to supply the `n_clusters` param (if you don't supply `n_clusters`, the algorithm will use `8` by default). You need to know that the optimal number of clusters differs case by case based on the dataset. K-Means can perform badly if the wrong number of clusters is used.

In advanced machine learning, data scientists try different numbers of clusters and evaluate the results with statistical measures (read [here](https://en.wikipedia.org/wiki/Cluster_analysis#External_evaluation)). We are not using statistical measures today but we'll use our eyes instead. In the cells below, experiment with different number of clusters and visualize with scatter plots. What number of clusters seems to work best for K-Means?

```python
# Your code here:
# Try different values for n_clusters and plot the results
for n_clusters in range(2, 6):
    kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(customers_scale)
    labels = kmeans.labels_
    data[f'Label_KMeans_{n_clusters}'] = labels

    plot(data['Detergents_Paper'], data['Milk'], labels, f'K-Means (n_clusters={n_clusters}): Detergents_Paper vs Milk')
```

# Bonus Challenge 3 - Changing DBSCAN `eps` and `min_samples`

Experiment changing the `eps` and `min_samples` params for DBSCAN. See how the results differ with scatter plot visualization.

```python
# Your code here
# Try different values for eps and min_samples and plot the results
for eps in [0.3, 0.5, 0.7]:
    for min_samples in [3, 5, 7]:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples).fit(customers_scale)
        labels = dbscan.labels_
        data[f'Label_DBSCAN_eps{eps}_min{min_samples}'] = labels

        plot(data['Detergents_Paper'], data['Milk'], labels, f'DBSCAN (eps={eps}, min_samples={min_samples}): Detergents_Paper vs Milk')
```
